```python

%matplotlib inline

import emupy as ep
import numpy as np
import matplotlib.pyplot as plt
import sklearn.gaussian_process as gp
import astropy.stats as astats
import scipy.linalg as la
from mpl_toolkits.mplot3d import Axes3D
import pickle as pkl
```

```python
# Load some training data
with open('lh_training.pkl','rb') as f:
    inp = pkl.Unpickler(f)
    dic = inp.load()
    data = dic['data']
    grid = dic['grid']
    
# Load cross-data
with open('cross_inspection.pkl','rb') as f:
    inp = pkl.Unpickler(f)
    dic = inp.load()
    data_cr = dic['data']
    grid_cr = dic['grid']
    # Sort the data
    fid_grid = np.median(grid_cr, axis=0)
    sort1 = np.where((grid_cr.T[1]==fid_grid[1])&(grid_cr.T[2]==fid_grid[2]))[0]
    sort2 = np.where((grid_cr.T[0]==fid_grid[0])&(grid_cr.T[2]==fid_grid[2]))[0]
    sort3 = np.where((grid_cr.T[0]==fid_grid[0])&(grid_cr.T[1]==fid_grid[1]))[0]
    data_cr1, grid_cr1 = data_cr[sort1], grid_cr[sort1]
    data_cr2, grid_cr2 = data_cr[sort2], grid_cr[sort2]
    data_cr3, grid_cr3 = data_cr[sort3], grid_cr[sort3]
    
rng = np.random.RandomState(1)
rando = rng.choice(np.arange(len(data)), replace=False, size=200)
select = np.array(np.zeros(len(data)), dtype=bool)
select[rando] = True
data_tr = data[select]
grid_tr = grid[select]
data_cv = data[~select]
grid_cv = grid[~select]
```

```python
fid_data = np.array(map(np.average,data_tr.T))
fid_grid = np.array(map(np.average,grid_tr.T))
```

```python
E = ep.Emu()
E.N_modes = 10
E.N_data = 33
E.N_params = 3
E.lognorm = False
E.cov_whiten = False
E.reg_meth = 'gaussian'
E.use_pca = True
```

```python
E.sphere(grid_tr, fid_grid = fid_grid, save_chol=True,norotate=True)

E.klt(data_tr,fid_data = fid_data,normalize=True)

# Set up gaussian process kwargs
kernel = gp.kernels.RBF(length_scale=7*np.ones(E.N_params),length_scale_bounds=(0.01,20)) + gp.kernels.WhiteKernel(noise_level=1e-4,)
optimizer='fmin_l_bfgs_b'
n_restarts_optimizer = 7
gp_kwargs = {'kernel':kernel,'optimizer':optimizer,'n_restarts_optimizer':n_restarts_optimizer }
E.gp_kwargs = gp_kwargs

E.train(data_tr,grid_tr,fid_grid = fid_grid, fid_data = fid_data)

E.predict(grid_cr1,sphere = True)
```

```python
pred = E.recon
```

```python
fig = plt.figure(figsize=(15,5))

w = 0

# Prints the graph "w" and the the next three for fun.
ax1 = fig.add_subplot(141)
ax1.scatter(grid_cr1[:,0],data_cr1[:,w])
ax1.plot(grid_cr1[:,0],pred[:,w])


ax1 = fig.add_subplot(142)
ax1.scatter(grid_cr1[:,0],data_cr1[:,w+1])
ax1.plot(grid_cr1[:,0],pred[:,w+1])

ax1 = fig.add_subplot(143)
ax1.scatter(grid_cr1[:,0],data_cr1[:,w+2])
ax1.plot(grid_cr1[:,0],pred[:,w+2])


ax1 = fig.add_subplot(144)
ax1.scatter(grid_cr1[:,0],data_cr1[:,w+3])
ax1.plot(grid_cr1[:,0],pred[:,w+3])
```
