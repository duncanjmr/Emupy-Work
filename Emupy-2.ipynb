```python
# The objective of this example is to show the most basic case of emupy use. We have two variables, x and y, and two outputs,
# f and g. Strategically, we have specified f and g to depend only on one of the two variables. The example data matrix goes as follows:
# 
#   coords   |         data
#  x1   y1       sin(x1)   0.5 * y1
#  x2   y2       sin(x2)   0.5 * y2
#  x3   y3       sin(x3)   0.5 * y3
#  x4   y4       sin(x4)   0.5 * y4
#
#
# In the actual processes, the coords and data matrices will be split up into two matrices, both n high and 2 wide.

# Import modules

%matplotlib inline

import emupy as ep
import numpy as np
import matplotlib.pyplot as plt
import sklearn.gaussian_process as gp
import astropy.stats as astats
import scipy.linalg as la
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Set parameters

E = ep.Emu()
E.N_modes = 2
E.N_data = 2
E.N_params = 2
E.lognorm = False
E.cov_whiten = False
E.reg_meth = 'gaussian'
use_pca = False
```

```python
# Set up data by defining the points along one axis, using that to create a meshgrid, and arranging the grid into an x and y
# column space of coordinates.
n=15
E.N_samples = n

axis = np.linspace(-1,1,n)
grid = np.meshgrid(axis,axis)

coords = np.array([grid[0].flatten(),grid[1].flatten()]).T


# Now we have a "coords" matrix of two columns and one entry for each coordinate in "grid". Now we create an identically sized
# data matrix with the following functions.
f = lambda x,y: np.sin(x*np.pi) + np.random.normal(0,1,1)/30 + 1
g = lambda x,y: y**2 + np.random.normal(0,1,1)/30

y1 = np.array(map(f,
                  coords[:,0],
                  coords[:,1])).T
y2 = np.array(map(g,
                  coords[:,0],
                  coords[:,1])).T
data = np.array([y1.flatten(),y2.flatten()]).T


# Finally, we create fid_coords, with the averages of each coordinate column, and fid_output, with the averages of each data column
fid_coords = np.array([np.median(coords[:,0]), np.median(coords[:,1])])
fid_data = np.array([np.median(data[:,0]), np.median(data[:,1])])
```

```python
# Now we start using emupy to analyze the data, with sphering and 

E.sphere(coords, fid_grid = fid_coords, save_chol=True,norotate=True)

E.w_norm = None

E.klt(data,fid_data = fid_data,normalize=False)

# Set up gaussian process kwargs
kernel = gp.kernels.RBF(length_scale=7*np.ones(E.N_params),length_scale_bounds=(0.01,20)) + gp.kernels.WhiteKernel(noise_level=1e-4,)
optimizer='fmin_l_bfgs_b'
n_restarts_optimizer = 10
gp_kwargs = {'kernel':kernel,'optimizer':optimizer,'n_restarts_optimizer':n_restarts_optimizer }
E.gp_kwargs = gp_kwargs

# Train the data
E.train(data,coords,fid_grid = fid_coords, fid_data = fid_data, compute_klt = False, use_pca=use_pca)

# create new coordinates for the predictions
nnew = 20
new_axis = np.linspace(-1,1,nnew)
pred_grid = np.meshgrid(new_axis,new_axis)
pred_coords = np.array([pred_grid[0].flatten(),pred_grid[1].flatten()]).T

E.predict(pred_coords,use_pca=use_pca,sphere=True)

pred_output = E.recon

pred_output = pred_output - np.array([fid_data]*len(pred_output))
```

```python
fig = plt.figure(figsize=(10,5))

# On axis one, plot the predicted surface of y1 and the points we used to calculate it.
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(pred_grid[0],pred_grid[1],pred_output[:,0].reshape(nnew,nnew),alpha=0.2,
                 rstride=2,
                 cstride=2)

ax1.scatter(grid[0],grid[1],y1.reshape(n,n))
ax1.set_title('Sine and fit',fontsize=14)


# Axis 2 is the same, except now we plot the fit of y2 and the points themselves.
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(pred_grid[0],pred_grid[1],pred_output[:,1].reshape(nnew,nnew),alpha=0.4,
                 rstride=1,
                 cstride=4)

ax2.scatter(grid[0],grid[1],y2.reshape(n,n),c='r',alpha=1)
ax2.set_title('Sinc and fit',fontsize=14)
fig.tight_layout()


# possible solution: specify minimum l as the smallest spacing between two points, max l as the domain of the variable. Like so:

#cipher_top = v_list + v_list[::-1]
```
