```python
%matplotlib inline

import emupy as ep
import numpy as np
import matplotlib.pyplot as plt
import sklearn.gaussian_process as gp
import astropy.stats as astats
import scipy.linalg as la
from mpl_toolkits.mplot3d import Axes3D
```

```python
E = ep.Emu()
E.N_modes = 1
E.N_data = 2
E.N_samples = n
E.N_params = 1
E.lognorm = False
E.cov_whiten = False
E.reg_meth = 'gaussian'
```

```python
n=20
E.N_samples = n

axis = np.linspace(-1,1,n)
grid = np.meshgrid(np.random.uniform(-1,1,n),np.random.uniform(-1,1,n))

f = lambda x,y: np.sinc(x*np.pi)
g = lambda x,y: np.sin(y*np.pi)

y1 = np.array(map(f,
                  grid[0].flatten(),
                  grid[1].flatten())).reshape(n,n)
y2 = np.array(map(g,
                  grid[0].flatten(),
                  grid[1].flatten())).reshape(n,n)

coors = np.array([grid[0].flatten(),grid[1].flatten()]).T
fid_coors = np.array([np.average(coors[:,0]),np.average(coors[:,1])])

output = np.array([y1.flatten(),y2.flatten()]).T
fid_output = np.array([np.average(output[:,0]),np.average(output[:,1])])

E.sphere(coors, fid_grid = fid_coors, save_chol=True,norotate=True)

E.klt(output,fid_data = fid_output,normalize=True)


kernel = gp.kernels.RBF(length_scale=np.ones(E.N_params)) + gp.kernels.WhiteKernel(noise_level=1e-6)
gp_kwargs = {'kernel':kernel}

E.gp_kwargs = gp_kwargs

E.train(coors,output,fid_grid = fid_coors, fid_data = fid_output, compute_klt = False, use_pca=False)




fitx = np.meshgrid(axis,axis)
fitx = np.array([fitx[0].flatten(),fitx[1].flatten()]).T
fitx = np.array(map(lambda x: x + np.array([0.01,0.01]),fitx))

E.predict(fitx,use_pca=True,output=True)

z = E.recon

fig = plt.figure(figsize=(10,5))

ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(grid[0],grid[1],z[:,1].reshape(n,n))



```
